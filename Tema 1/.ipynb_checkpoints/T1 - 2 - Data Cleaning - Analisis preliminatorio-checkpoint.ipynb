#### Borrado de valores que faltan

```python
data.dropna(axis = 0, how = "all")
# Axis = 0 es el eje X, se ejecuta dropna si todas (all) cumplen el criterio de ser nulas
```

```python
data.dropna(axis = 0, how = "any")
```

```python
data2 = copy.copy(data)
```

```python
data2["body"] = data2["body"].fillna(0)
data2["home.dest"] = data2["home.dest"].fillna("Unknown")
data.head(3)
```

```python
data2["age"].fillna(method = "ffill")
#Introduce el conocido anteriormente
```

```python
data2["age"].fillna(method = "backfill")
#Introduce el que conoce posteriormente
```

```python
data2["age"] = data2["age"].fillna(data2["age"].mean())
```

```python
def createDummies(df, var_name):
    dummy = pd.get_dummies(df[var_name], prefix = var_name)
    df = df.drop([var_name], axis = 1)
    df = pd.concat([df, dummy], axis = 1)
    return df
```

```python
data3 = copy.copy(data2)
```

```python
createDummies(data3, "sex")
```
