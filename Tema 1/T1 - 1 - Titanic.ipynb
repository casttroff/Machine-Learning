# Carga de datos con open

```python
data3 = open(path + "/" "customer-churn-model/Customer Churn Model.txt", "r")
```

```python
data3.seek(0)
```

```python
cols = data3.readline().strip().split(",")
```

```python
cols
```

```python
n_cols = len(cols)
```

```python
n_cols
```

```python
counter = 0 
main_dict = {}
for col in cols:
    main_dict[col] = []
main_dict
```

```python
for line in data3:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter += 1
        
```

```python
df3 = pd.DataFrame(main_dict)
```

```python
df3.head()
```

```python
infile = path + "/" "customer-churn-model/Customer Churn Model.txt"
outfile = path + "/" "customer-churn-model/Tab Customer Churn Model.txt"
```

```python
with open(infile, "r") as infile1:
    with open(outfile, "w") as outfile1:
        for line in infile1:
            fields = line.strip().split(",")
            outfile1.write("\t".join(fields))
            outfile1.write("\n")
```

```python
df4 = pd.read_csv(outfile, sep = "\t")
```

# Leer datos desde URL

```python
import csv, urllib3
medals_url = "http://winterolympicsmedals.com/medals.csv"
pathmedal="/Users/User/Desktop/Machine-Learning/datasets/Medals/medals.csv"
```

```python
http = urllib3.PoolManager()
r = http.request('GET', medals_url)
r.status
response = r.data.decode("utf-8")
medals = response.strip().split("\n")
with open(pathmedal,"w") as medalfile:
    for medal in medals:
        medalfile.write(medal)
        medalfile.write("\n")
medalfile1 = pd.read_csv(pathmedal)
medalfile1.to_json(path + "/Medals/medals.json")
#medalfile1
```
